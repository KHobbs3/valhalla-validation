```python
import pandas as pd
```

```python
routes = pd.read_csv("Jordan's_routes_analysis.csv", header=[0])
```

```python
routes.columns = routes.iloc[0]

routes = routes.iloc[1:].reset_index(drop=True)

routes.head()
```

```python
routes_filter = routes[['observation','Class_1','Class_2','Cost','Cost/dist']]
```

```python
routes_filter.dtypes
```

```python
routes_filter["Cost"] = pd.to_numeric(routes_filter["Cost"] )
routes_filter["Cost/dist"] = pd.to_numeric(routes_filter["Cost/dist"] )
```

```python
routes_filter.columns
```

```python
routes_filter.groupby(['Class_1','Class_2']).median().sort_values('Cost')
```

```python
cost_ott = pd.read_csv('costs-ott_analysis.csv')
```

```python
cost_ott.columns = cost_ott.iloc[0]

cost_ott = cost_ott.iloc[1:].reset_index(drop=True)

cost_ott.head()
```

```python
cost_ott["cost"] = pd.to_numeric(cost_ott["cost"] )
cost_ott["cost/dist"] = pd.to_numeric(cost_ott["cost/dist"] )
cost_ott["dur/dist"] = pd.to_numeric(cost_ott["dur/dist"] )
cost_ott["cost/dur"] = pd.to_numeric(cost_ott["cost/dur"] )
```

```python
cost_ott.groupby(['Class_1','Class_2']).median().sort_values('cost')
```
